```python
def key_dialogue(data):
    keys = []
    for i in range((data.shape[0])):
        try:
            source_id = list(data['source_id'])[i]
            abbr = source_dict[source_id]
            key = abbr+str(list(data['N_dialogue'])[i])
            keys.append(key)
        except Exception:
            keys.append('')
    data['key_dialogue'] = keys
```

```python
import re
```

```python
# from Stack Overflow

def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
                                u"\U00000000-\U00000009"
                                u"\U0000000B-\U0000001F"
                                u"\U00000080-\U00000400"
                                u"\U00000402-\U0000040F"
                                u"\U00000450-\U00000450"
                                u"\U00000452-\U0010FFFF"
                                "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
```

```python
clean_texts = [deEmojify(str(text)) for text in vkdata['post_text']]
vkdata['clean_post_text'] = clean_texts
```

```python
def pair_dialogs(data):
    paired_dialogues = []
    keys = list(data['key_dialogue'].unique())
    for k in keys:
        dialog_data = data.loc[data['key_dialogue'] == k]
        if len(dialog_data) == 1:
            dialog = [dialog_data.iloc[0]['clean_post_text'], dialog_data.iloc[0]['clean_text']]
            paired_dialogues.append(dialog)
        else:
            dpairs = []
            for i in range(len(dialog_data)):
                try:
                    pair = [dialog_data.iloc[i]['clean_text'], dialog_data.iloc[i+1]['clean_text']]
                    paired_dialogues.append(pair)
                except IndexError:
                    continue
    return paired_dialogues
```

```python
%%time
paired_dialogues = pair_dialogs(vkdata)
```

```python
f = open('paired_dialogues.json')
json_file = json.load(f)
```

```python
new_paired_dialogues = json_file['paired_dialogues']
```

```python
%%time
for item in new_paired_dialogues:
    phrase1, phrase2 = item # item - list, phrases - elements
    vk_appeal = re.compile(pattern = '\[id[0-9]+\|([A-Za-z–ê-–Ø–∞-—è–Å—ë]+)\], ')
    to_change = re.compile(pattern = '\[id[0-9]+\|[A-Za-z–ê-–Ø–∞-—è–Å—ë]+\], ')
    
    phr1 = vk_appeal.search(phrase1)
    if phr1:
        name1 = str(phr1.group(1)) + ', '
        phrase_1 = to_change.sub(name1, phrase1)
        item[0] = phrase_1
    else:
        pass
    
    phr2 = vk_appeal.search(phrase2)
    if phr2:
        name2 = str(phr2.group(1)) + ', '
        phrase_2 = to_change.sub(name2, phrase2)
        item[1] = phrase_2
    else:
        pass
```
